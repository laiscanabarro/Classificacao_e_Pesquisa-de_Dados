```python
import pandas as pd
```

```python
def processar_nome(nome):
    nomes_separados = nome.split(' ')  # Separe os nomes por vírgula e espaço
    nomes_filtrados = [n for n in nomes_separados if len(n) > 2]  # Remova os nomes com ponto
    return '%20'.join(nomes_filtrados)  # Junte os nomes formatados novamente
```
